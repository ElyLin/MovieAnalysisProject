```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#first we must import the libraries that we want to use 
```

```python
movie_budgets_df = pd.read_csv('zippedData/tn.movie_budgets.csv.gz') #Michael has expertise on this dataframe 
movie_studios = pd.read_csv('zippedData/bom.movie_gross.csv.gz')     #Teigen's has expertise on this dataframe
title_basics = pd.read_csv('zippedData/imdb.title.basics.csv.gz')    #Louis' has expertise on this dataframe
title_ratings = pd.read_csv('zippedData/imdb.title.ratings.csv.gz')  #as well as this one 
tmdb = pd.read_csv('zippedData/tmdb.movies.csv.gz')                  #Ely has expertise on this dataframe
```

# Now that we have loaded all of our raw data, let's merge it together!

```python
#firstly, the tmbd database has it's genre's encoded so we must decode the genre's using the below dictionary
genres={'28':"Action",
'12':"Adventure",
'16':"Animation",
'35':"Comedy",
'80':"Crime",
'99':"Documentary",
'18':"Drama",
'10751':"Family",
'14':"Fantasy",
'36':"History",
'27':"Horror",
'10402':"Music",
'9648':"Mystery",
'10749':"Romance",
'878':"Science Fiction",
'10770':"TV Movie",
'53':"Thriller",
'10752':"War",
'37':"Western"}

tmdb = tmdb.replace(pd.Series(genres).astype(str), regex=True)
tmdb['genre_ids']=tmdb['genre_ids'].str.strip('[]').str.split(', ')
```

```python
#secondly, we must merge the two IMDB dataframes
title_master = title_basics.merge(title_ratings, how='inner')
```

```python
#thirdly, we must merge the tmdb database with the imdb database. we chose to do an inner merge, here because we 
#are only interested in the movies that these dataframes have in common
tmdb_and_imdb = tmdb.merge(title_master, how = 'inner')
tmdb_and_imdb
```

```python
#here we must merge the Box Office Mojo data with the TMDB and IMDB dataframes, and again we are only interested in 
#the movies that are shared between those dataframes so we are doing an inner merge
tmdb_and_imdb_and_bomo = tmdb_and_imdb.merge(movie_studios, how = 'inner')
```

```python
#next, because we are going to be using the fincial data from The Numbers, we need to drop the domestic gross and 
#foreign gross columns as it currently exists
without_finacials = tmdb_and_imdb_and_bomo.drop(['domestic_gross', 'foreign_gross'], axis = 1)
without_finacials
```

```python
#the next dataframe we want to fold in has the column "movie" but we want to rename it to title so the merge can 
#happen on the title of the film, so we must rename the column
movie_budgets_df = movie_budgets_df.rename(columns={'movie':'title'})
```

```python
#now we want to merge the fincial dataframe from The Numbers with all the others based on Title, so we specify that 
#with the on = 'title'
master_unclean = without_finacials.merge(movie_budgets_df, on = 'title', how = 'inner')
master_unclean
```

```python
#let's take a look at the amount of information we have now and the types of information in our columns
#we can see where we have duplicate columns, we now have x and y versions to distinguish them from each other
#runtime_minutes and genres are the only columns that do not have 16665 non null values after this merge
master_unclean.info()
```

```python
#now we want to exclude movies from this master dataframe which are less than half an hour and more than three 
#hours, because we want to base our recomendations on films that are typically seen in the movie theater, no 
#short film stand alones, and no art-film style epics, and put that information on a new dataframe called run_time
run_time = master_unclean.loc[(master_unclean['runtime_minutes'] > 30) & (master_unclean['runtime_minutes'] < 180)]
run_time
```

```python
#here, we want to drop some columns where the information is irrelivant or duplicated 
less_columns = master_unclean.drop(['Unnamed: 0', 'id_x', 'original_title', 'tconst', 'start_year', 'id_y',\
                                    'genres', 'release_date_x', 'primary_title'], axis = 1)
```

```python
#let's take a look at where our dataframe is right now
less_columns
```

```python
#because we want our reccomendations to be relevent to the current market, we only want to look at movies from the 
#last few years, so let's filter those out
less_columns=less_columns.loc[less_columns['year'] >= 2016]
```

```python
#now we don't need the year column since the remaining release date information we will be using is the month taken 
#from release_date_y
less_columns=less_columns.drop(['year'], axis=1)
```

```python
#let's take a look at what we have at this point
less_columns
```

```python
#now we must clean up this financial information by removing punctuation and making sure that these numbers are 
#integers and therefore we will be able to perform mathmatical manipulation on them in the future 
less_columns['production_budget'] = less_columns['production_budget'].str.replace(',', '')
less_columns['production_budget'] = less_columns['production_budget'].str.replace('$', '')
less_columns['production_budget'] = less_columns['production_budget'].astype(int)
less_columns['domestic_gross'] = less_columns['domestic_gross'].str.replace(',', '').str.replace('$', '')
less_columns['domestic_gross'] = less_columns['domestic_gross'].astype(int)
less_columns['worldwide_gross'] = less_columns['worldwide_gross'].str.replace(',', '').str.replace('$', '')
less_columns['worldwide_gross'] = less_columns['worldwide_gross'].astype(int)
less_columns['total_gross'] = less_columns['domestic_gross'] + less_columns['worldwide_gross']
less_columns['ROI'] = (less_columns['total_gross'] / less_columns['production_budget']) * 100
```

```python
#now let's make a Return on Investment column, using the fincial data we just cleaned
less_columns['ROI'] = (less_columns['total_gross'] / less_columns['production_budget']) * 100
```

# Now that we have cleaned data, let's analyze it a bit so we can provide reccomendations

```python
#let's sort by studio---FILL IN LATER WITH TEIGEN'S WORK
less_columns.groupby(['studio'], sort = True).max()
```

```python
#to explore the trends by genre, first we must "explode" the genre_ids so that each film is listed once for each of 
#the genres it is in 
exploded = less_columns.explode('genre_ids', ignore_index=False)
```

```python
#let's reset our index for this genre data so that we can use iloc should we need to
exploded.reset_index()
```

```python
#let's also make a new column and put just the month when that movie was released from release_date_y
exploded['month'] = exploded['release_date_y'].str[0:3]
```

```python
#now let's take another look at our exploded dataframe
exploded
```

```python
#let's also assign this month column to "month" so we can call upon this series later
month = exploded['month']
```

```python
#let's make a dataframe which only has the month, the genre, the ROI, and the title, for ease of use
month_genre_roi_title = exploded.drop(['release_date_y','original_language', 'popularity', 'vote_average', 'vote_count',\
                                      'runtime_minutes', 'averagerating', 'numvotes', 'studio', \
                                      'production_budget', 'domestic_gross', 'worldwide_gross',\
                                      'total_gross'], axis=1)
```

```python
#this is much eaiser to look at and figure out what is happening
month_genre_roi_title
```

```python
#let's make a new dataframe where it is clear what the mean Return on Investment is for each genre
genre_roi = month_genre_roi_title.groupby(['genre_ids']).mean().sort_values('ROI', ascending=False)
genre_roi
```

```python
#what are the genres with the top five largest means for Return on Investment? 
top_5_genres=list(genre_roi.index[0:5])
top_5_genres
```

```python
month_genre_roi_title=month_genre_roi_title[month_genre_roi_title['genre_ids'] in top_5_genres]
month_genre_roi_title
```
